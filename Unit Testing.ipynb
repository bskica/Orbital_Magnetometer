## Peakfinder Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
from mypeakfinder import peakfinder

data = np.random.rand(50)
x = np.linspace(0,50, 50)
p = peakfinder(data)
px = []
for i in range(len(p)):
    for j in range(len(data)):
        if p[i] == data[j]:
            px.append(x[i])
plt.plot(x, data)

plt.plot(px, p)
```
